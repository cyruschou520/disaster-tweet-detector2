```python
# ============================
# Final Year Project - Disaster Tweet Classification
# UTAR FYP - NLP Approach
# ============================

# STEP 1: Import Libraries
import pandas as pd
import re
import emoji
import nltk
import spacy
import torch
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from transformers import AutoTokenizer, AutoModel, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset

# ============================
# STEP 2: Load Dataset
# ============================
file_path = "tweets.csv"   # ✅ Kaggle disaster tweet dataset
df = pd.read_csv(file_path)

print("Dataset shape:", df.shape)
print("Columns in dataset:", df.columns.tolist())
print(df.head())

# ============================
# STEP 3: Preprocessing Tweets
# ============================

# Download NLTK stopwords (only first time)
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define preprocessing function
def preprocess_tweet(text):
    text = str(text).lower()                                 # Lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)      # Remove URLs
    text = re.sub(r"@\w+", "", text)                         # Remove mentions
    text = re.sub(r"#", "", text)                            # Remove hashtag symbols
    text = emoji.demojize(text)                              # Convert emojis to words
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc 
              if token.is_alpha and token.text not in stop_words]
    return " ".join(tokens)

# Apply preprocessing
df["clean_text"] = df["text"].astype(str).apply(preprocess_tweet)
print(df[["text", "clean_text"]].head())

# ============================
# STEP 4: Feature Extraction
# ============================

# ---- Option A: Classical NLP (TF-IDF) ----
vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(df["clean_text"])
y = df["target"]   # ✅ Kaggle's label column

print("TF-IDF shape:", X_tfidf.shape)

# ---- Option B: Deep Learning (BERT Embeddings) ----
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-uncased")
bert_model = AutoModel.from_pretrained("bert-base-multilingual-uncased")

# Example encoding for 2 tweets
sample_texts = df["clean_text"].iloc[:2].tolist()
encoded = tokenizer(sample_texts, padding=True, truncation=True, return_tensors="pt")
with torch.no_grad():
    embeddings = bert_model(**encoded).last_hidden_state.mean(dim=1)

print("BERT embeddings shape:", embeddings.shape)

# ============================
# STEP 5: Model Building
# ============================

# ---- A: Baseline Logistic Regression (with TF-IDF) ----
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("\nLogistic Regression Results (TF-IDF):")
print(classification_report(y_test, y_pred))

# ---- B: Fine-tuning BERT for Classification ----

# Convert dataset to HuggingFace Dataset
dataset = Dataset.from_pandas(df[["clean_text", "target"]])

# Rename target -> labels (Trainer requirement)
dataset = dataset.rename_column("target", "labels")

# Tokenize dataset
def encode_batch(batch):
    return tokenizer(batch["clean_text"], truncation=True, padding="max_length", max_length=64)

dataset = dataset.map(encode_batch, batched=True)

# Remove the raw text column (not needed for training)
dataset = dataset.remove_columns(["clean_text"])

# Train-test split
train_test = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = train_test["train"]
test_dataset = train_test["test"]

# Load model
bert_classifier = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-multilingual-uncased", num_labels=2
)

# Training setup
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=1,     # 🔁 Increase to 3-5 for better results
    weight_decay=0.01,
    logging_dir="./logs"
)

# Trainer
trainer = Trainer(
    model=bert_classifier,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

# Train model
trainer.train()

# Evaluate
results = trainer.evaluate()
print("\nBERT Fine-tuning Results:", results)

```
